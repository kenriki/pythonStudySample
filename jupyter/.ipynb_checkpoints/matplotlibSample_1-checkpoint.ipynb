```python
import matplotlib.pyplot as plt
import numpy as np

# データ生成
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100) 

# プロット
plt.plot(x, y, label="test")

# 凡例の表示
plt.legend()

# グリッド線を表示するならTrue
plt.grid(True)

# プロット表示(設定の反映)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcdefaults()
fig, ax = plt.subplots()

# データ
subjects = np.array(['HP', '攻撃', '防御', '特攻', '特防', '素早さ'])
scores = np.array([80, 82, 83, 100, 100, 80])

# グリッド線を表示するならTrue
plt.grid(True)

#棒グラフの作成
ax.barh(subjects, scores, align='center')

#グラフの軸の設定
ax.set_yticks(subjects, labels=subjects, fontname="MS Gothic")
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('各能力値', fontname="MS Gothic")
ax.set_title('フシギバナ', fontname="MS Gothic")
```
