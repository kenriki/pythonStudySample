```python
import matplotlib.pyplot as plt
import numpy as np

# データ生成
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100) 

# プロット
plt.plot(x, y, label="test")

# 凡例の表示
plt.legend()

# グリッド線を表示するならTrue
plt.grid(True)

# プロット表示(設定の反映)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path


def main():
    plt.rcdefaults()
    fig = plt.figure(figsize = (13,8), facecolor="lightgray", tight_layout=True)
    ax = fig.subplots(1, 3)
    
    csvfile = Path(
        "C:\\Users\\user\\Desktop\\作業\\branch\\python\\jupyter\\pokemon_data.csv")
    # CSVファイルの読み込み、配列に格納
    pkmn_data = pd.read_csv(
        csvfile, header=0, encoding="utf-8").values.tolist()
    #pkmn_data = pd.read_csv(csvfile, index_col=[0],encoding="utf-8",header=None).to_string(index=False).values.tolist()
    # Windowsの特殊文字が使われている場合は,encoding="cp932"

    #print(np.array(pkmn_data))

    # データ
    subjects = np.array(['HP', '攻撃', '防御', '特攻', '特防', '素早さ'])
    names = np.array(['フシギバナ', 'リザードン', 'カメックス'])
    #scores = np.array([80, 82, 83, 100, 100, 80])
    
    index = 0
    for itemScoreName in names:
        print("インデックス：" + str(index) + ", 値：" + itemScoreName)
        ax[index].set_title(itemScoreName, fontname="MS Gothic")
        # 棒グラフコール
        printGraph(ax[index], subjects, pkmn_data[index])
        # グラフの軸の設定
        ax[index].set_xlabel('各能力値', fontname="MS Gothic")
        ax[index].set_yticks(subjects, labels=subjects, fontname="MS Gothic")
        ax[index].invert_yaxis()  # labels read top-to-bottom
        # グリッド線を表示するならTrue
        ax[index].grid(True)
        index += 1

    
    

# 棒グラフの作成
def printGraph(ax, subje, data):
    print(data)
    ax.barh(subje, data)


if __name__ == "__main__":
    main()
```
